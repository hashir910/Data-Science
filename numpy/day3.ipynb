```python
import pandas as pd
import numpy as np

narr=np.array([1.0,2.1,3.32,4.99,5.10])
narr
```

```python
# arthimetic

print(narr+10)
print(narr*10)
print(narr/10)
print(narr**10)
print(narr-10)
```

```python
# built in methods
print(narr)
print(np.sqrt(narr))
print(np.round(narr))
print(np.floor(narr))
print(np.ceil(narr))
```

```python
#vectorized methods
narr2=np.array([6,7,8,9,10])

print(np.round(narr)+narr2)
print(np.round(narr)-narr2)
print(np.round(narr)/narr2)
print(np.round(narr)*narr2)
print(np.round(narr)**narr2)
```

```python
# comparision operator

print(narr>10)
print(narr<10)
print(narr==10)
print(narr!=10)
```
