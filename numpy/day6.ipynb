```python
# random integers------------------

rang=np.random.default_rng()
print(rang.integers(1,5))
print(rang.integers(1,101))
print(rang.integers(1,101,size=3))
print(rang.integers(1,101,size=(3,4)))
```

```python
# random choice-------------------

rng=np.random.default_rng()
arr2=np.array([1,2,3,4,5])
choic=rng.choice(arr2)
choic=rng.choice(arr2,size=3)
choic=rng.choice(arr2,size=(3,2))
print(choic)
```

```python
# shuffle

rng.shuffle(arr2)
arr2
```
