**aggregate functions**

```python
arr3=np.array([[1,2,3,4]])
arr4=np.array([[1],[2],[3],[4],
              [1],[2],[3],[4]])

print(arr3.sum())
print(arr4.mean())
print(arr4.std())
print(arr4.var())
```

```python
print(np.sum(arr3))
print(np.sum(arr4,axis=0))
print(np.sum(arr4,axis=1))
print(np.mean(arr4,axis=0))
print(np.mean(arr4,axis=1))
print(np.std(arr4,axis=1))
print(np.std(arr4,axis=0))
```
