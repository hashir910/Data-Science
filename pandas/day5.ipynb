```python
import pandas as pd
import numpy as np

coffee=pd.read_csv('../warmup-data/coffee.csv')
bios=pd.read_csv('../data/bios.csv')
```

```python
bios['last_name']=bios['name'].str.split(' ').str[-1]
bios
```

```python
bios['birth_day']=pd.to_datetime(bios['born_date']).dt.day_name()
bios.drop(columns=['birth_year'],inplace=True)
```

```python
def height(row):
    if row['height_cm']>180:
        return 'tall'
    else:
        if row['height_cm']<160:
            return 'short'
        else:
            return 'medium'

bios['height_category']=bios.apply(height,axis=1)
bios
```

```python
bios['timespan']=pd.to_datetime(bios['died_date'])-pd.to_datetime(bios['born_date'])

bios.dropna(how='any',inplace=True)
bios
```
