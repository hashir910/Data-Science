```python
import pandas as pd
import numpy as np

df=pd.read_csv('books/messy_data.csv')
col=[]
for i in df.columns:
    col.append(i.strip(' '))
df.columns=col
```

```python
# for id

df['Emp ID']=pd.Series(range(101,101+len(df)))
df

df['Full Name']=df['Full Name'].replace(' ',np.nan)
df
```

```python
#fixing age column


# df['age']=df['age'].str.replace(r'[A-Za-z]|[^A-Za-z0-9\s]','',regex=True)
# df['age'] = df['age'].replace(['?', '', ' ', 'NA', 'n/a'], np.nan)
# df['age']=df['age'].astype('float')
# df['age']=df['age'].fillna(df['age'].mean())
# df['age']=df['age'].astype('int')
# df

df['age'] = (
    df['age']
    .astype(str)
    .str.replace(r'[^0-9]', '', regex=True)   # Keep only digits
    .replace(['', ' ', 'NA', 'n/a', '?'], np.nan)  # Replace invalids
    .astype(float)                            # Convert to float
)

# Fill NaNs with mean, convert to int
df['age'] = df['age'].fillna(df['age'].mean()).astype(int)

df
```

```python
#for gender

df['Gender']=(
    df['Gender']
    .str.strip()
    .str.lower()
    .replace({
        'm':'Male', 'male':'Male',
        'f':'Female', 'female':'Female',
        '?':'Unknown'

    })
    )

df


```

```python
#fixing dept
df['Dept']=(
    df['Dept']
    .str.strip()
    .str.upper()
    .replace(
        '',np.nan
    )
)
df
```

```python
#fixing salary
df['Salary$']=(
    df['Salary$']
    .str.strip(' ')
    .str.replace(r'[Kk]$','000',regex=True)
    .str.replace(r'[^\d.]','',regex=True)
    .replace('',np.nan)
    .astype(float)
    
)

df['Salary$'].fillna(round(df['Salary$'].mean()),inplace=True)
df
```

```python
#fixing experience
wordtn={
    'one':1,
    'two':2,
    'three':3,
    'four':4,
    'five':5,
    'six':6,
    # '':np.nan

}
df['Experience(Yrs)']=(
    df['Experience(Yrs)']
    .str.strip()
    .str.lower()
    .replace(wordtn)
    .astype(float)
)
df.isna().sum()
df['Experience(Yrs)']=df['Experience(Yrs)'].fillna(df['Experience(Yrs)'].mean()).astype(int)

# df['Experience(Yrs)']=df['Experience(Yrs)']

df.fillna('unknown',inplace=True)
df

```

```python
# fixing date
import pandas as pd
import re
from dateutil import parser
import re

def clean_date_text(x):
    if pd.isna(x) or str(x).strip() == '' or str(x).lower() == 'unknown':
        return None
    x = re.sub(r'[./]', '-', str(x).strip())  # unify separators
    x = re.sub(r'\s+', '-', x)
    return x

clean= df['JoinDate'].apply(clean_date_text)


def parse_messy_date(x):
    if x is None:
        return pd.NaT
    try:
        return parser.parse(x, dayfirst=True, yearfirst=False)
    except Exception:
        return pd.NaT

final_date= clean.apply(parse_messy_date)
df['JoinDate']=final_date
df['JoinDate']=df['JoinDate'].fillna(pd.Timestamp('2021-01-01'))





df
```

```python
#fixing cities



def cleaning_cityfunc(x):
    
    if pd.isna(x) or str(x).strip()=='':
        return 'Unknown'
    
    if bool(re.match(r'^[A-Za-z]', str(x))):
        first_word=x[0].upper()
        return(first_word)


df['cleaned_city']=df['City'].apply(cleaning_cityfunc).replace({
            'K':'KHI','I':'ISL','L':'LHR','U':'Unknown'
        })
df['City']=df['cleaned_city']
df.drop(columns=['cleaned_city'],inplace=True)
```

```python
kbd=df.loc[df.groupby('Dept')['Experience(Yrs)'].idxmax()]

finalanalyze=kbd[['Emp ID','Full Name','Dept','age','Gender','Experience(Yrs)']]
finalanalyze.to_csv('messyanalyzedata1.csv', index=False)






```
