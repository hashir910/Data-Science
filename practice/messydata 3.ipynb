```python
import pandas as pd
import numpy as np

df=pd.read_csv('messy_data.csv')
df
```

```python
# for id

df['Emp ID']=pd.Series(range(101,101+len(df)))
df

```

```python
#fixing age column


# df['age']=df['age'].str.replace(r'[A-Za-z]|[^A-Za-z0-9\s]','',regex=True)
# df['age'] = df['age'].replace(['?', '', ' ', 'NA', 'n/a'], np.nan)
# df['age']=df['age'].astype('float')
# df['age']=df['age'].fillna(df['age'].mean())
# df['age']=df['age'].astype('int')
# df

df['age'] = (
    df['age']
    .astype(str)
    .str.replace(r'[^0-9]', '', regex=True)   # Keep only digits
    .replace(['', ' ', 'NA', 'n/a', '?'], np.nan)  # Replace invalids
    .astype(float)                            # Convert to float
)

# Fill NaNs with mean, convert to int
df['age'] = df['age'].fillna(df['age'].mean()).astype(int)

df
```

```python
#for gender

df['Gender ']=(
    df['Gender ']
    .str.strip()
    .str.upper()
    .replace({
        'm':'Male',
        'f':'Female',
        '?':'Unknown'

    })
    )

df


```

```python
#fixing datedf
df['JoinDate'] = pd.to_datetime(df['JoinDate'], errors='coerce', dayfirst=False)


df
```
